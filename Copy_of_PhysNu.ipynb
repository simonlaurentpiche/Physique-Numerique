```python
## Init

import numpy as np
import scipy
from scipy.special import factorial
from matplotlib import pyplot as plt
```

```python
## Ex 2.11 

## a2

n = int(input("Enter a value for n "))
k = int(input("Enter a value for k "))

def binomial(n,k):
    if n >=0 and k>=0 and n>=k:  
      a,b,c = factorial(n), factorial(k), factorial(n-k)
      return int(a/(b*c))
    else: 
      print("Enter only positives values for n and/or k and n needs to be higher than k")

binomial(n,k)



```

```python
## b
N = int(input("Enter a value for N for the Pascal's Triangle "))


for n in range(1,N+1):
  triangle = []
  triangle.append(binomial(1,0))
  for i in range(1,n+1):
    triangle.append(binomial(n,i))
  print(triangle)

```

```python
## c

toss = int(input("Nombre de lancers "))
heads = int(input("Nombre de Heads "))
sum=0

print("Probabilit√© d'obtenir " + str(heads) + " heads sur " str(toss) + " lancers. ")
print(binomial(toss, heads)/(2**toss))
print("Probabilit√© d'obtenir " + str(heads) + " heads et plus sur " + str(toss) + " lancers. ")
for i in range(toss-heads):
  sum += binomial(toss, i+1)/(2**toss)
print(sum)
```

```python
data = np.loadtxt('millikan.txt', delimiter = "\t" )
#print(data)

x = []
y = []
for i in range(len(data)):
  x.append(data[i,0]), y.append(data[i,1])

def functions(x, y):
  Ex = np.sum(x)/len(x)
  Ey = np.sum(y)/len(y)
  Exx = np.sum(np.multiply(x,x))/len(x)
  Exy = np.sum(np.multiply(x,y))/len(x)
  return Ex, Ey, Exx, Exy

def slope(Ex, Ey, Exx, Exy):
  m = (Exy - Ex*Ey)/(Exx-Ex**2)
  c = (Exx*Ey - Ex*Exy)/(Exx - Ex**2)
  return m,c
plt.scatter(x,y);


print(functions(x,y))
m, c = slope(*functions(x,y))

print(m,c)
avg = np.multiply(x,m) + c

print(avg)


plt.plot(x, avg)


print(m*1.602e-19)
```
